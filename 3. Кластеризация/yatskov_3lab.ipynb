```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('google_review_ratings.csv')
```

```python
df
```

```python
df.info()
```

Видим, есть NaN значения и столбцы с типом данных object.

```python
df.describe()
```

```python
# Удалим последний столбец в котором только 2 непустые строки
df.drop('Unnamed: 25', axis=1, inplace=True)
```

```python
df.info()
```

```python
# Удалим столбец с id пользователей, тк это (грубо говоря) дублирование номеров строк, и информация уникальна для каждой строки 
df.drop('User', axis=1, inplace=True)
```

```python
# Удалим все пустые значения, их всего 2 в категории 12 и 24

df.dropna(inplace=True)
```

```python
df.info()
```

```python
# Удалаять какие либо категории не имеет смысла, тк именоо на их основе и производится кластеризация
# Так же все данные уже нормализованны в шкале от 0 до 5
df['Category 11']
```

```python
#Несмотря на то, что в столбце числовая информация, конвертируем его в float
df['Category 11'] = pd.to_numeric(df['Category 11'],errors = 'coerce')
df['Category 11']
```

```python
X = df
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn import metrics
```

```python
a = 2
b = 10
inertia = []
silh = []
for k in range(a, b):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    inertia.append(np.sqrt(kmeans.inertia_))
    silh.append(metrics.silhouette_score(X, kmeans.predict(X)))

plt.plot(range(a, b), inertia, marker='s');
plt.xlabel('количество кластеров')
plt.ylabel('сумма расстояний до центроидов');
silh
```

```python
# Оптимальные k = 6 или 7, но значенния силлуэта равны 0.15
# так же заметно что при увеличенни кол-ва кластеров увеличивается значение силлуэта
```

```python
kmeans_50 = KMeans(n_clusters=50, random_state=1).fit(X)
```

```python
metrics.silhouette_score(X, kmeans_50.predict(X))
```

```python
# однако даже на 50ти кластерах значение силуэта достигает только 0.20
```

```python
print(list(kmeans_50.labels_))
```

```python
# Так же видно, что кластеризация все же работает более менее адекватно и не запихивает все значения в один кластер
# Вероятно всему виной очень близкое и кучное расположение значений
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# попробуем использовать Агломеративную кластеризацию
```

```python
a = 2
b = 10
silh = []
for k in range(a, b):
    agc = AgglomerativeClustering(n_clusters=k, linkage='ward')
    silh.append(metrics.silhouette_score(X, agc.fit_predict(X)))

silh
```

```python
# Результат получился еще хуже, чем при использовании метода к средних.
# Таким образом, данные модели кластеризации не могут однозначно разделить данные на кластеры. 
```
